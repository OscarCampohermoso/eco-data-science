# Base de datos

Con un software de "base de datos" usualmente queremos realizar las siguientes operaciones:

- Leer una base de datos
- Creación de nuevas columnas("variables")
- Filtrar filas
- Filtrar columnas
- Tablas resúmen
- Cruce de datos
- Unión de basede datos
- Grabar una base de datos

## Objetos importantes de la librería Pandas

**1. DataFrame**

**2. Series**

**3. GroupBy**

```python
import pandas as pd
```

```python
file = 'carteras.txt'
df = pd.read_csv(file, sep=';')
df.head()
```

```python
pd.Series([1,2,3,4,5])
```

```python
df2 = df.set_index('Run Fondo')

df2
```

```python
df.columns
```

```python
basex = df2.loc[[8011, 10500], ['FFM_6010100', 'FFM_6010211']] # [filas, columnas] esto slicing se llama "fancy indexing"

basex
```

```python
basex.columns
```

```python
basex = basex.rename(columns={'FFM_6010100': 'Nemo', 'FFM_6010211': 'Montos'})
basex
```

```python
basex['MontosMM'] = basex['Montos'] / 1000
basex
```

```python
agg = basex.groupby('Nemo').agg({'MontosMM': 'sum'})
agg
```
