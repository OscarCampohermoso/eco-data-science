```python
# Compras de productos XXL que se hicieron durante el mes de Junio.

query = 'SELECT *, SUBSTR(Date, 1,1) AS "Mes"\
         FROM "Ventas Amazon"\
         WHERE Mes = "6" AND Size = "XXL"'

df = pd.read_sql(query, conn)

df
```

```python
# Consulta para ver si la orden fue hecha por una empresa o un consumidor, en la ciudad de MUMBAI

query = 'SELECT "Order ID",\
             CASE\
                 WHEN B2B = 1 THEN "B2B - Empresa"\
                 WHEN B2B = 0 THEN "B2C - Consumidor"\
                 ELSE "Desconocido"\
             END AS TipoCliente\
         FROM "Ventas Amazon"\
         WHERE "ship-city" = "MUMBAI"\
         ORDER BY TipoCliente'

df = pd.read_sql(query, conn)

df
```

```python
# CTE = Common table expression

# Genera una consulta donde primero se cree una CTE, que guarde todas las ordenes de las tallas "L", cuyo monto individual haya sido
# mayor a 1000, para despues hacer una consulta sobre dicha CTE, que me muestre las ordenes que no envÃ­o Amazon (es decir las entrego el vendedor)

query = 'WITH CTE AS (\
          SELECT "Order ID", Date, Fulfilment, Size, Amount\
          FROM "Ventas Amazon"\
          WHERE Size = "L" AND Amount > 1000\
          )\
         SELECT *\
         FROM CTE\
         WHERE Fulfilment = "Merchant"'

df = pd.read_sql(query, conn)

df
```

```python
# Consultas anidadas 
# 1. Transacciones cuyo amount haya sido mayor que el promedio de todas las ventas

query = 'SELECT *\
         FROM "Ventas Amazon"\
         WHERE "Amount" > (SELECT AVG("Amount") FROM "Ventas Amazon")'

df = pd.read_sql(query, conn)

df
```

```python
# Consultas anidadas 
# 2. Montos totales de todas las tallas que sean mas grandes (o igual) de "L", que sean mayores de 1.000.000 de rupias y sus cantidad sea mayor que 15000

query = 'SELECT "Cantidad", "Monto total", "Size"\
         FROM (SELECT COUNT(*) as Cantidad, "Size", SUM("Amount") as "Monto total" FROM "Ventas Amazon" GROUP BY "Size")\
         WHERE "Cantidad" > 15000 AND "Monto total" > 1000000 AND ("Size" = "L" OR "Size" = "XL" OR "Size" = "XXL" OR "Size" = "3XL" OR "Size" = "4XL" OR "Size" = "5XL" OR "Size" = "6XL")'

df = pd.read_sql(query, conn)

df
```
