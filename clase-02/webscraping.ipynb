# Beautiful Soup

```python
import csv
from bs4 import BeautifulSoup
import pandas as pd
import lxml
```

```python
url = "https://www.worldometers.info/world-population/population-by-country/"
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text)
```

```python
soup
```

```python
data = soup.find_all("table")[0]
```

```python
df_population = pd.read_html(str(data))[0] # Convertimos la tabla en un dataframe
df_population.head() # Mostramos las primeras 5 filas
```

```python
export_csv = df_population.to_csv(r"le chemin\nom du fichier.csv", index = None, header = True)
```
