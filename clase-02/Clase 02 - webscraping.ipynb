## Ambientes de trabajo

Un ambiente virtual (o environment) es un entorno virtual que le permite al desarrollador establecer específicamente qué versiones de software y librerías quiere utilizar para algún proyecto particular. 

En nuestra clase, el ambiente nos permitirá a todos tener las mismas versiones y evitar confusiones o errores.

#### Gestión del ambiente

conda env create $\rightarrow$ para crear el ambiente

conda env update $\rightarrow$ para actualizar el ambiente

#### Pasos al programar

Cada vez que queramos programar, debemos:

1. Abrir Anaconda Prompt o el Terminal
2. Activar el ambiente: **conda activate eae253** 
3. Abrir un jupyter notebook: **jupyter notebook**
