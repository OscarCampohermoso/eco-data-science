```python
def conteo_votos(data):
    d = {}
    
    for mesa, votos_c1, votos_c2 in data:
        if mesa in d:
            d[mesa] = d[mesa][0] + votos_c1, d[mesa][1] + votos_c2
        else:
            d[mesa] = votos_c1, votos_c2
    return d
```

```python
def mis_votaciones(data):
    d = {}

    for m, v1, v2 in data:
        if m in d:
            d[m][0] += v1
            d[m][1] += v2
        else:
            d[m] = [v1, v2]
    return d
```

```python
def resultados(data):
    suma1 = 0
    suma2 = 0
    for m, v1, v2 in data:
        suma1 += v1
        suma2 += v2
    return suma1, suma2
```

```python
def verificar_mesa(data):
    lista_si_estan = []
    lista_no_estan = []
    for m, _, _ in data:
        lista_si_estan.append(m)
    
    for i in range(0, 351):
        if i not in lista_si_estan:
            lista_no_estan.append(i)
```

```python
def mesas_faltantes(data):
    mesas_reportadas = []
    for mesa, _, _ in data:
        if mesa not in mesas_reportadas:
            mesas_reportadas.append(mesa)

    mesas_faltantes = []
    for i in range(1, 351):
        if i not in mesas_reportadas:
            mesas_faltantes.append(i)

    return mesas_faltantes
```

```python
mesas_faltantes(data)
```

```python
verificar_mesa(data)
```

```python
resultados(data)
```

```python
mis_votaciones(data)
```

```python
conteo_votos(data)
```

```python
def numero_de_votos_por_mesa(data):
    d = {}
    for mesa, votos_c1, votos_c2 in data:
        if mesa in d:
            d[mesa] += votos_c1 + votos_c2
        else:
            d[mesa] = votos_c1 + votos_c2
    return d
```

```python
numero_de_votos_por_mesa(data)
```

```python
def votos_por_mesa(data):
    votos_mesa = {}
    
    for mesa, votos_A, votos_B in data:
        if mesa in votos_mesa:
            votos_mesa[mesa] = (votos_mesa[mesa][0] + votos_A, votos_mesa[mesa][1] + votos_B)
        else:
            votos_mesa[mesa] = (votos_A, votos_B)
    
    return votos_mesa
```

```python
votos_por_mesa(data)
```
